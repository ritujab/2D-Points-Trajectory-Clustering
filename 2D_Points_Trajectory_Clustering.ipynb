```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Define the synthetic dataset
data = np.array([[1, 1], [1, 2], [1, 3], [2, 2], [2, 3], [3, 4]])

# Define the number of trajectories
n_trajectories = 3

# Define the number of points in each trajectory
n_points = 2

# Reshape the data into a list of trajectories
trajectories = []
for i in range(n_trajectories):
    start = i * n_points
    end = start + n_points
    trajectory = data[start:end]
    trajectories.append(trajectory)
trajectories = np.array(trajectories)

# Now run K-means clustering on the trajectories
kmeans = KMeans(n_clusters=n_trajectories)
kmeans.fit(trajectories.reshape((n_trajectories, n_points*2)))

# Add labels assigned to each trajectory by the clustering algorithm
labels = kmeans.labels_

# Finally we plot the trajectories
colors = ['red', 'green', 'blue']
for i in range(n_trajectories):
    x = trajectories[i, :, 0]
    y = trajectories[i, :, 1]
    plt.plot(x, y, c=colors[labels[i]])
plt.show()
```
